## Importing the libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
import math
```

## Importing the dataset

```python
data_df=pd.read_csv("Social_Network_Ads.csv")
data_df.head()
```

## Select Features and Target Variable



```python
x=data_df.iloc[:,2:4].to_numpy()
y=data_df.iloc[:,4].to_numpy()
x=x.astype('float')
y=y.astype('float')
```

## Splitting the dataset into the Training set and Test set

```python
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.15,random_state=42,shuffle=True)

print(x_train.shape,x_test.shape,y_train.shape,y_test.shape)
```

## Feature Scaling

```python
def P(x,a,b):
    g=np.matmul(x,a)+b
    for i in range(0,len(g)):
        g[i]=math.exp(g[i])
    h=1/(1+g)
    return h
```

```python
def loss(x,a,b,y):
    Prob=P(x,a,b)
    l=0.0
    for i in range(0,len(y)):
        
        l=l+y[i]*(math.log(Prob[i])) + (1-y[i])*(math.log(1-Prob[i]))
    return l
```

```python
def gradient(x,a,b,y):
    m,n=x.shape
    Prob=P(x,a,b)
    g=(y-Prob)[:, np.newaxis]
    da= np.sum(g*x,axis=0)
    db=np.sum(y-Prob)
    assert(da.shape==(n,))
    return (da,db)
```

```python
def gradient_ascent(x,y,learning_rate=0.01,epochs = 10):
    m,n=x.shape
    a=np.zeros(n,)
    loss_m=[]
    b=0

    for i in range(epochs):
        
        l=loss(x,a,b,y)
        loss_m.append(l)
        da,db=gradient(x,a,b,y)
        a=a+da*learning_rate
        b=b+db*learning_rate
        print("Epoch ",i+1," Completed!","likelihood = ",l)
    
    print("Training completed!!")
    assert(a.shape==(n,))
    return (loss_m,a,b)



```

## Training the Logistic Regression model on the Training set

```python
epochs = 200             
learn_rate = 0.01          

liklihood,a,b = gradient_ascent(x_train,y_train,learn_rate,epochs)
```
